```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
START_STAT = 300
MAX_STAT = 2000
atk, mag = np.meshgrid(np.arange(START_STAT,MAX_STAT),np.arange(START_STAT,MAX_STAT))
print(atk.shape)
print(mag.shape)
deff = [50, 100, 200, 400, 600, 800]
spr = 100
```

```python
%matplotlib notebook

for i in range(0,len(deff)) :
    atk_mag_damage = 0.5*((atk**2/deff[i]) + (mag**2)/spr)
    plt.subplot(3,3,i+1)
    plt.imshow(atk_mag_damage, vmin=0)
    plt.title('DEF = ' + str(deff[i]))
    plt.gca().set_xticks([])
    plt.xlabel('ATK')
    plt.gca().set_yticks([])
    plt.ylabel('MAG')
    plt.colorbar()
plt.tight_layout()
```
